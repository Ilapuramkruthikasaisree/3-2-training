```python
import json #built in python librarry
#json.dumps() --> python data to json format api
#json.loads() --> json strings to python
#json.dump() -->python file --> json file
#json.load() -->json string file -->python file
data = {
    "email":"mru@gmail.com",
    "password":"mruh"
}
print(data)
type(data)
```

```python
result=json.dumps(data)
print(result)
print(type(result))
```

```python
result2=json.loads(result)
print(result2)
print(type(result2))
```
